# Importing Libraries

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

# Creating Data

```python
x = np.linspace(0,50,501)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
df = pd.DataFrame(data=y,index=x,columns=['Sine'])
df
```

```python
df.head()
```

```python
len(df)
```

```python
test_percent = 0.1
len(df)*test_percent
```

```python
test_point = np.round(len(df)*test_percent)
test_point
```

```python
test_ind = int(len(df)-test_point)
```

```python
train = df.iloc[:test_ind]
```

```python
test = df.iloc[test_ind:]
```

```python
train
```

```python
test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(train)
```

# Train Test Split and Time Generator Creation

```python
scaled_train = scaler.transform(train)
scaled_test = scaler.transform(test)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
# help(TimeseriesGenerator)
```

```python
length = 25
batch_size = 1
generator = TimeseriesGenerator(scaled_train,scaled_train,
                                length=length,batch_size=batch_size)
```

# Creating A Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,SimpleRNN,LSTM
```

## Working With Simple RNN

```python
n_features = 1
```

```python
model = Sequential()

model.add(SimpleRNN(50,input_shape=(length,n_features)))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.summary()
```

```python
model.fit_generator(generator,epochs=5)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.head()
```

```python
plt.plot(losses)
```

```python
first_eval_batch = scaled_train[-length:]
first_eval_batch = first_eval_batch.reshape((1,length,n_features))
```

```python
model.predict(first_eval_batch)
```

```python
scaled_test[0]
```

```python
test_predictions = []
first_eval_batch = scaled_train[-length:]
current_batch = first_eval_batch.reshape((1,length,n_features))

for i in range(len(test)):
    current_pred = model.predict(current_batch)[0]
    test_predictions.append(current_pred)

    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
true_predictions = scaler.inverse_transform(test_predictions)
```

```python
# true_predictions
```

```python
test['Predictions'] = true_predictions
test.head()
```

```python
test.plot(figsize=(12,8))
```

## Working With LSTM and Callbacks

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss',patience=2)
```

```python
length = 49
generator = TimeseriesGenerator(scaled_train,scaled_train,
                                length = length, batch_size = 1)
validation_gen = TimeseriesGenerator(scaled_test,scaled_test,
                                     length=length,batch_size=1)
```

```python
model = Sequential()

model.add(LSTM(50,input_shape=(length,n_features)))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(generator,epochs=20,validation_data=validation_gen,callbacks=[early_stop])
```

```python
losses_lstm = pd.DataFrame(model.history.history)
losses_lstm.head()
```

```python
losses_lstm.plot()
```

```python
test_predictions = []
first_eval_batch = scaled_train[-length:]
current_batch = first_eval_batch.reshape((1,length,n_features))

for i in range(len(test)):
    current_pred = model.predict(current_batch)[0]
    test_predictions.append(current_pred)

    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
true_predictions = scaler.inverse_transform(test_predictions)
test['LSTM Predictions'] = true_predictions
test.head()
```

```python
test.plot(figsize=(12,8))
```

## Taking Full Data

```python
full_scaler = MinMaxScaler()
scaled_df = full_scaler.fit_transform(df)
```

```python
generator = TimeseriesGenerator(scaled_df,scaled_df,length = length,batch_size=batch_size)
```

```python
model = Sequential()

model.add(LSTM(50,input_shape=(length,n_features)))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
forecast = []
first_eval_batch = scaled_df[-length:]
current_batch = first_eval_batch.reshape((1,length,n_features))

for i in range(25):
    current_pred = model.predict(current_batch)[0]
    forecast.append(current_pred)

    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```
